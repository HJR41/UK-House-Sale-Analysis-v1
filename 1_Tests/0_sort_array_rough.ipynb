```python
array = [520, 569, 521, 2, 4, 3, 20, 141, 639, 699, 5, 8, 2, 5, 999, 52, 4, 87, 984, 2,489, 36, 1, 25, 65, 78, 972]
```

```python
def sort_array(array):
    if len(array) <= 1: # if length of array is 1 or less than, it is already sorted, so just return array.
        return array
    else:
        centre = array[len(array) // 2] # else, find the centre of the array to compare the rest of the list to, you could use any value here. for eg array[1] or array[-1]
        left = [i for i in array if i < centre] # if i in list is smaller than centre, append to left
        middle = [i for i in array if i == centre] # if i == centre append to middle
        right = [i for i in array if i > centre] # if i > centre, append to right
        return sort_array(left) + middle + sort_array(right) # recursively pass the new lists (left & right (middle is already sorted)) back through the function until len(array) <1, then left + middle + right to get a sorted list.
```

```python
def sort_array_2(array):
    left = []
    middle = []
    right = []
    if len(array) <=1:
        return array
    else:
        centre = array[len(array) // 2]
        for i in array:
            if i < centre:
                left.append(i)
            else:
                if i == centre:
                    middle.append(i)
                else:
                    right.append(i)
    return sort_array_2(left) + middle + sort_array_2(right)
```
